
This s a demo network analysis of commecting the streets using NetworkX in Python. This uses fake data.
My thinking: Using this network analysis connect the streets and once we have a prediction of which streets may be unsafe after the quake we can find the shortest ditance between street from parent location to school location.



```
import networkx as nx
```

```
#Symmetric STreets
G_symmetric = nx.Graph()
G_symmetric.add_edge('Street1','Street2')
G_symmetric.add_edge('Street1','Street3')
G_symmetric.add_edge('Street1','Street4')
G_symmetric.add_edge('Street1','Street5')
G_symmetric.add_edge('Street2','Street3')
G_symmetric.add_edge('Street2','Street4')
G_symmetric.add_edge('Street2','Street5')
G_symmetric.add_edge('Street5','Street3')
```

```
nx.draw_networkx(G_symmetric)
```

```
#Assymmetric Streets
G_asymmetric = nx.DiGraph()
G_asymmetric.add_edge('A','B')
G_asymmetric.add_edge('A','D')
G_asymmetric.add_edge('C','A')
G_asymmetric.add_edge('D','E')
```

```
nx.spring_layout(G_asymmetric)
```

```
nx.draw_networkx(G_asymmetric)
```

```
#the weights can be how unsafe or safe the roads would be.
G_symmetric = nx.Graph()
G_symmetric.add_edge('Street1','Street2',weight=25)
G_symmetric.add_edge('Street1','Street3', weight=8)
G_symmetric.add_edge('Street1','Street4', weight=11)
G_symmetric.add_edge('Street1','Street5',weight=1 )
G_symmetric.add_edge('Street2','Street3', weight=4)
G_symmetric.add_edge('Street2','Street4', weight=7)
G_symmetric.add_edge('Street2','Street5',weight=1)
G_symmetric.add_edge('Street5','Street3', weight=1)
```

```
nx.draw_networkx(G_symmetric)
```

```
nx.degree(G_symmetric, "Street4")
```

```
#Local Clustering Coefficient
nx.average_clustering(G_symmetric)
```

```
nx.shortest_path(G_symmetric, 'Street4', 'Street5')
```

```
T = nx.bfs_tree(G_symmetric, 'Street5')
```

```
nx.draw_networkx(T)
```
